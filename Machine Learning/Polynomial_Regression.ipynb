```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Importing datasets
dataset =pd.read_csv('E:\\Machine Learning A-Z Template Folder\\Part 2 - Regression\\Section 6 - Polynomial Regression\\Position_Salaries.csv')
print(f'data set\n{dataset}\n')

X = dataset.iloc[:,1:2].values
Y = dataset.iloc[:,2].values

#Fitting model to linear Regration
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)


#Fitting model to Polynomial Regration
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
print(poly_reg)

lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,Y)

#Visualising Linear Regration
plt.scatter(X,Y, color='red')
plt.plot(X,lin_reg.predict(X))
plt.title('Trust vs Bluff(LinearRegression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()

#Visualising Linear Polynomial
X_grid = np.arange(min(X), max(X), 0.1)
#print(f'Before shape{X_grid.shape}')
X_grid = X_grid.reshape(len(X_grid), 1)
#print(f'after shape{X_grid.shape}')

plt.scatter(X,Y, color='blue')
plt.plot(X_grid,lin_reg2.predict(poly_reg.fit_transform(X_grid)))    
plt.title('Trust vs Bluff(PolynomialRegression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
